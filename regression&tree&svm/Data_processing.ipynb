```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('sph6004_assignment1_data.csv')
```

```python
data
```

```python
# classify aki as with or without aki
# 1 represents aki positive 0 represents aki negative
data.loc[data['aki'] >= 1, 'aki'] = 1
```

```python
# quantify gender
dict_gender = {'F': 0, 'M': 1}
data['gender'] = data['gender'].replace(dict_gender)
```

```python
# quantify race
race = data['race'].value_counts()
dict_race = {key: i for i, key in enumerate(race.index)}
data['race'] = data['race'].replace(dict_race)
```

```python
# keep columns with missing value less than 0.1
missing_ratio = data.isna().mean()
features_to_keep = missing_ratio[missing_ratio <= 0.1].index
data = data[features_to_keep]
```

```python
data_clean = data.fillna(value=data.mean(), axis=0)
```

```python
aki_1 = (data_clean['aki'] == 1).sum()
aki_2 = (data_clean['aki'] == 0).sum()
print(f'{aki_1} aki equal to 1, {aki_2} equal to 0')
```

```python
# because of the imbalanced data
# randomly drop half of aki_positive
indexes_to_drop = data_clean.index[data_clean['aki'] == 1]
np.random.seed(42)
random_indexes_to_drop = np.random.choice(indexes_to_drop, size=len(indexes_to_drop)//2, replace=False)
data_clean = data_clean.drop(random_indexes_to_drop)
data_clean = data_clean.reset_index(drop=True)
```

```python
aki_1 = (data_clean['aki'] == 1).sum()
aki_2 = (data_clean['aki'] == 0).sum()
print(f'{aki_1} aki equal to 1, {aki_2} equal to 0')
```
